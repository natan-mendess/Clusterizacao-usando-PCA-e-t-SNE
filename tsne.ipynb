Imports

```python
import numpy as np # Algebra linear
import pandas as pd # Processar dataset
import matplotlib.pyplot as plt # Plotar gr√°fico
import seaborn as sns # Plotar heatmap
from sklearn.preprocessing import StandardScaler # Reescalar valores do dataset
from openTSNE import TSNE
```

Ler/tratar database

```python
df = pd.read_csv("dataset.csv")
df['Target']=df['Target'].apply(lambda x: 1 if x=='Graduate' else 0)
scaler=StandardScaler()
scaled_data = scaler.fit_transform(df)
```

t-SNE

```python
tsne = TSNE(n_components=2, perplexity=500)
x_tsne = tsne.fit(scaled_data)

plt.figure(figsize=(8,6))
plt.scatter(x_tsne[:,0], x_tsne[:,1], c=df['Target'],cmap='plasma')
plt.xlabel('Componente tsne 1')
plt.ylabel('Componente tsne 2')
```
